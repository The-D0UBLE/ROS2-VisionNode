```python
import os
import shutil
from pathlib import Path
from ultralytics import YOLO
from ultralytics.data.converter import convert_coco

path = "/home/double/Documents/school/yolo_testV2/"
source = "/home/double/Documents/school/yolo_testV2/coco_data/cocv8n-seg_data/"
output = f"{path}yolo-dataset/"


if not(os.path.exists(output)):
    convert_coco(labels_dir=source + "train/", save_dir=output + "train/" ,use_segments=True)
    convert_coco(labels_dir=source + "valid/", save_dir=output + "valid/" ,use_segments=True)

```

```python
image_exts = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.tif', '.gif')


# Copy train images
for file_path in Path(source + "train/").glob("*"):
    if file_path.suffix.lower() in image_exts:
        dest_path = os.path.join(output + "train/images", file_path.name)
        if not os.path.exists(dest_path):
            shutil.copy(file_path, dest_path)
            print(f"Copied: {file_path.name}")
        else:
            print(f"Skipped (already exists): {file_path.name}")

# Copy valid images
for file_path in Path(source + "valid/").glob("*"):
    if file_path.suffix.lower() in image_exts:
        dest_path = os.path.join(output + "valid/images", file_path.name)
        if not os.path.exists(dest_path):
            shutil.copy(file_path, dest_path)
            print(f"Copied: {file_path.name}")
        else:
            print(f"Skipped (already exists): {file_path.name}")

```

```python
# Copy train images
for file_path in Path(output + "train/labels/_annotations.coco/").glob("*"):
        dest_path = os.path.join(output + "train/labels", file_path.name)
        if not os.path.exists(dest_path):
            shutil.move(file_path, dest_path)
            print(f"Copied: {file_path.name}")
        else:
            print(f"Skipped (already exists): {file_path.name}")

# Copy valid images
for file_path in Path(output + "valid/labels/_annotations.coco/").glob("*"):
        dest_path = os.path.join(output + "valid/labels", file_path.name)
        if not os.path.exists(dest_path):
            shutil.move(file_path, dest_path)
            print(f"Copied: {file_path.name}")
        else:
            print(f"Skipped (already exists): {file_path.name}")
```
